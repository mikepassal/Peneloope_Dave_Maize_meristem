```python
def read_in_and_get_coeff_of_var(file_name, ranktype="reg"):
    current_file = '/data/johlee/expData/maize/' + file_name
    result = pyreadr.read_r(current_file)
    pd_result = result['metaExpData']
    ranked_maize_test = pd_result.rank()
    ranked_coef_variance = ranked_maize_test.var(axis = 1)/ranked_maize_test.mean(axis =1)
    
    sum_maize = pd_result.sum(axis = 1)
    zero_in_expt = pd_result.loc[sum_maize==0].index
    ranked_coef_variance.loc[zero_in_expt] = np.nan
    
    # num_expts = pd_result.shape[1] - pd_result[pd_result>0].isnull().sum(axis =1)
    # ranked_coef_variance.loc[num_expts<10] = np.nan

    return ranked_coef_variance
```

```python
ranked_coef_df_all_expt = pd.DataFrame(index = pd_result.index,)

for file in maize_data_files:
    ranked_coef_df_all_expt[file] = read_in_and_get_coeff_of_var(file)
```
